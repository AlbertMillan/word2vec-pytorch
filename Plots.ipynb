```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_art = 10
emb_dim = 100
```

```python
loss_hist_a = np.load("results/loss_" + str(num_art) + "__" + str(emb_dim) + ".npy")
spear_hist_a = np.load("results/spear_" + str(num_art) + "__" + str(emb_dim) + ".npy")

loss_hist_b = np.load("results/loss_" + str(num_art*10) + "__" + str(emb_dim) + ".npy")
spear_hist_b = np.load("results/spear_" + str(num_art*10) + "__" + str(emb_dim) + ".npy")

loss_hist_c = np.load("results/loss_" + str(num_art*1000) + "__" + str(emb_dim*2) + ".npy")
spear_hist_c = np.load("results/spear_" + str(num_art*1000) + "__" + str(emb_dim*2) + ".npy")
```

```python
base_art = 10
emb_dim = 100

for i in range(1,5):
    num_art = base_art ** i
    
    if num_art != 1000:
        
        
        loss_hist_a = np.load("results/loss_" + str(num_art) + "__" + str(emb_dim) + ".npy")
        spear_hist_a = np.load("results/spear_" + str(num_art) + "__" + str(emb_dim) + ".npy")
        
        loss_hist_b = np.load("results/loss_" + str(num_art) + "__" + str(emb_dim*2) + ".npy")
        spear_hist_b = np.load("results/spear_" + str(num_art) + "__" + str(emb_dim*2) + ".npy")
        
        loss_hist_c = np.load("results/loss_" + str(num_art) + "__" + str(emb_dim*3) + ".npy")
        spear_hist_c = np.load("results/spear_" + str(num_art) + "__" + str(emb_dim*3) + ".npy")
        
        plt.figure()
        plt.plot(np.arange(len(loss_hist_a)), loss_hist_a, label='100_emb')
        plt.plot(np.arange(len(loss_hist_b)), loss_hist_b, label='200_emb')
        plt.plot(np.arange(len(loss_hist_c)), loss_hist_c, label='300_emb')
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
#         plt.ylim(0,None)
        plt.title("Model Loss " + str(num_art) + " art.")
        plt.legend()
        plt.savefig("graphs/loss_art_" + str(num_art) + ".png", bbox_inches="tight")
        plt.show()
        
        
        plt.figure()
        plt.plot(np.arange(len(spear_hist_a)), spear_hist_a, label='100_emb')
        plt.plot(np.arange(len(spear_hist_b)), spear_hist_b, label='200_emb')
        plt.plot(np.arange(len(spear_hist_c)), spear_hist_c, label='300_emb')
        plt.xlabel("Epochs")
        plt.ylabel("Spearman Coefficient")
#         plt.ylim(0,6000)
        plt.title("Spearman Coeff. " + str(num_art) + " art.")
        plt.legend()
        plt.savefig("graphs/spear_art_" + str(num_art) + ".png", bbox_inches="tight")
        plt.show()
        
        print("Spear " + str(num_art) + " articles " + str(emb_dim) + ":" + str(max(spear_hist_a)) )
        print("Spear " + str(num_art) + " articles " + str(emb_dim*2) + ":" + str(max(spear_hist_b)) )       
        print("Spear " + str(num_art) + " articles " + str(emb_dim*3) + ":" + str(max(spear_hist_c)) )       
```

```python
base_emb_dim = 100
num_art = 10

for i in range(1,4):
    emb_dim = base_emb_dim * i
    
    if num_art != 1000:
        
        
        loss_hist_a = np.load("results/loss_" + str(num_art) + "__" + str(emb_dim) + ".npy")
        spear_hist_a = np.load("results/spear_" + str(num_art) + "__" + str(emb_dim) + ".npy")
        
        loss_hist_b = np.load("results/loss_" + str(num_art*10) + "__" + str(emb_dim) + ".npy")
        spear_hist_b = np.load("results/spear_" + str(num_art*10) + "__" + str(emb_dim) + ".npy")
        
        loss_hist_c = np.load("results/loss_" + str(num_art*1000) + "__" + str(emb_dim) + ".npy")
        spear_hist_c = np.load("results/spear_" + str(num_art*1000) + "__" + str(emb_dim) + ".npy")
        
        plt.figure()
        plt.plot(np.arange(len(loss_hist_a)), loss_hist_a, label='10_art')
        plt.plot(np.arange(len(loss_hist_b)), loss_hist_b, label='100_art')
        plt.plot(np.arange(len(loss_hist_c)), loss_hist_c, label='10000_art')
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
#         plt.ylim(0,None)
        plt.title("Model Loss " + str(emb_dim) + " emb. dim.")
        plt.legend()
        plt.savefig("graphs/loss_emb_" + str(emb_dim) + ".png", bbox_inches="tight")
        plt.show()
        
        
        plt.figure()
        plt.plot(np.arange(len(spear_hist_a)), spear_hist_a, label='10_art')
        plt.plot(np.arange(len(spear_hist_b)), spear_hist_b, label='100_art')
        plt.plot(np.arange(len(spear_hist_c)), spear_hist_c, label='10000_art')
        plt.xlabel("Epochs")
        plt.ylabel("Spearman Coefficient")
#         plt.ylim(0,6000)
        plt.title("Spearman Coeff. " + str(emb_dim) + " emb. dim.")
        plt.legend()
        plt.savefig("graphs/spear_emb_" + str(emb_dim) + ".png", bbox_inches="tight")
        plt.show()
```
